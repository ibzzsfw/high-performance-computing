```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# data: Dimension,GPU_mean,CPU_mean,GPU_std,CPU_std
df = pd.read_csv('./result.csv')
```

```python
# plot with error bars
fix, ax = plt.subplots()
ax.errorbar(df['Dimension'], df['CPU_mean'], yerr=df['CPU_std'], fmt='o')
ax.errorbar(df['Dimension'], df['GPU_mean'], yerr=df['GPU_std'], fmt='o')
ax.set_xlabel('Dimension')
ax.set_ylabel('Time (ms)')
ax.set_title('Matrix Multiplication')
ax.legend(['CPU', 'GPU'])
plt.show()
```
